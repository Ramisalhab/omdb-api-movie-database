Write a query that selects a specific movie/series based on its title

```python
#Test your Zaka
# Query to find a movie with the title "Friends"
result = collection.find_one({"Title": "Friends"})

# Print the result
if result:
    print(result)
else:
    print("Movie not found.")
```

Write a query that returns all movies/series that were released after 2006.

```python
#Test your Zaka
results = collection.find({"Year": {"$gt": "2006"}})

# Print the results
for result in results:
    print(result)
```

Write a query that can find comedy series/movies only.

```python
#Test your Zaka
# Query to find all comedy series/movies
results = collection.find({"Genre": {"$regex": "Comedy", "$options": "i"}})

# Print the results
for result in results:
    print(result)
```
